```python
Type_new = pd.Series([]) 
# running a for loop and asigning some values to series 
for i in range(len(df)): 
    if df["Team"][i] == "Grass": 
        Type_new[i]="Green"
  
    elif df["Team"][i] == "Fire": 
        Type_new[i]="Orange"
  
    elif df["Team"][i] == "Water": 
        Type_new[i]="Blue"
  
    else: 
        Type_new[i]= df["Team"][i] 
  
          
# inserting new column with values of list made above         
df.insert(2, "Team_New", Type_new) 
  
# list output 
df.head() 
```
